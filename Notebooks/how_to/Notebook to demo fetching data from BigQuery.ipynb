# Using BigQuery with python

### Fetching from BigQuery

Define query

```python
query = """
SELECT 
    *
FROM `gamesys-eu-dev-sbtech-data.sb_tech_presentation.BetSnapshot` 
WHERE lastSettlementDate = "2023-01-15"
LIMIT 100
"""
```

Use BigQuery function to make call.

```python
bq_output_df = bq_query(query)
bq_output_df.head(5)
```

### Persisting to BigQuery

Filter dataframe to not include columns with dict structures. For simplicity, filtering for a df with only the betID column
NB: this step needed as having difficulty persisting columns with nested data

```python
bq_output_to_persis_df = bq_output_df[['betID']]
```

Persist data to bq_persist

```python
bq_persist(
    input_df = bq_output_to_persis_df, 
    table_name = 'JamesL.test_table',
    if_exists = 'replace'
)
```
